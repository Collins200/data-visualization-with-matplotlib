matplotlib visualization

```python
import matplotlib.pyplot as plt
import numpy as np
# the x values/plots will take default values
# mfc stand for marker face color and ms for marker size
# ls stands for line style
# c stands for line color
y=np.array([0,20,40,70,30,50])
plt.plot(y,marker='o',linestyle='--',markersize=20,markerfacecolor='c',color="green",linewidth='10')
plt.title('simple visualization')
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.show()
```

```python
# marker description
# ”.”	point
# ”,”	pixel
# “o”	circle
# “v”	triangle_down
# “^”	triangle_up
# “<”	triangle_left
# “>”	triangle_right
# “1”	tri_down
# “2”	tri_up
# “3”	tri_left
# “4”	tri_right
# “8”	octagon
# “s”	square
# “p”	pentagon
# “*”	star
# “h”	hexagon1
# “H”	hexagon2
# “+”	plus
# “x”	x
# “D”	diamond
# “d”	thin_diamond
# “|”	vline
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([10,60,40,30,90])
xpoints=np.array([0.1,0.3,0.6,0.7,0.9])
plt.plot(xpoints,ypoints,ls=''or'',marker="*",mfc='r',ms=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([10,40,30,90])
y2points=np.array([20,50,10,70])
xpoints=np.array([0.1,0.6,0.7,0.9])
x2points=np.array([0.2,0.4,0.5,0.9])
plt.plot(xpoints,ypoints,x2points,y2points)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.suptitle("GRAPHS") #govern the whole graphs
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1, 2, 1)
plt.title('graph1') #title labelling comes after subplot
plt.plot(x,y)
# another plot
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1, 2, 2)
plt.title('graph2')
plt.plot(x,y)
plt.show()

#subplot takes the argument number of rows,number of columns and then number of plots 
```
