matplotlib visualization

```python
import matplotlib.pyplot as plt
import numpy as np
# the x values/plots will take default values
# mfc stand for marker face color and ms for marker size
y=np.array([0,20,40,70,30,50])
plt.plot(y,marker='o',ms=20,mfc='m')
plt.title('simple visualization')
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.show()
```
